```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
```

```python
lsoas = gpd.read_file(r'Access_to_Healthy_Assets_and_Hazards_AHAH_E08000012/data/Access_to_Healthy_Assets_and_Hazards_AHAH/Local_Authority_Districts/E08000012/shapefiles/E08000012.shp')
```

```python
ahah_data = pd.read_csv(r'Access_to_Healthy_Assets_and_Hazards_AHAH_E08000012/data/Access_to_Healthy_Assets_and_Hazards_AHAH/Local_Authority_Districts/E08000012/tables/E08000012.csv')
```

```python
lsoas.head()
```

```python
ahah_data.head()
```

```python
ahah = lsoas.join(ahah_data.set_index('lsoa11cd'), on='lsoa11cd')
```

```python
ahah.head(3)
```

```python
f, ax =plt.subplots(1, figsize=(10,10))
ahah.plot(column='ahah',
          scheme='equal_interval',
         k=5, ax=ax, legend=True)
```

```python
f, ax =plt.subplots(1, figsize=(10,10))
ahah.plot(column='ahah',
         scheme = 'quantiles',
         k=5, ax=ax, legend=True, edgecolor='k', linewidth=.1)
```

```python
f, ax = plt.subplots(1, figsize=(10,10))
ahah.plot(column='ahah',
         scheme = 'FisherJenks',
         k=5, ax=ax, legend=True, edgecolor='k', linewidth=.1)
```

```python

mn = ahah['ahah'].mean()
ahah['classi'] =  ahah['ahah'].apply(lambda x: 'above' if (x > mn) else 'below' )
```

```python
import matplotlib.colors
cm = matplotlib.colors.LinearSegmentedColormap.from_list("", ['#8dd3c7','#bebada'])
```

```python
f, ax = plt.subplots(1, figsize=(10,10))
ahah.plot( ax=ax,
column='classi', cmap=cm, legend=True, edgecolor='k', linewidth=0.2)
```
