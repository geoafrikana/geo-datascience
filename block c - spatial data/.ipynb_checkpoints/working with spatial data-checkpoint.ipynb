## Task I: Tweak your map
With those three layers, try to complete the following tasks:
- Make a map of the Liverpool neighborhoods that includes the following characteristics:
    - Features a title
    - Does not include axes frame
    - It has a figure size of 10 by 11
    - Polygons are all in color "#525252" and 50% transparent
    - Boundary lines (“edges”) have a width of 0.3 and are of color "#B9EBE3"

## Task 2: Non-spatial manipulations
Focus on the LSOA liv layer and use it to do the following:
1. Calculate the area of each neighbourhood
2. Find the five smallest areas in the table. Create a new object (e.g. smallest with them only)
3. Create a multi-layer map of Liverpool where the five smallest areas are coloured in red, and the rest appear in black.

```python
smallest = liv.area.sort_values().head(6)
smallest
```

```python
smallest_places = liv[liv.area <= smallest.iloc[4]]
```

```python
fig, ax = plt.subplots(1, figsize=(10,11))
plt.title("Map of Liverpool showing Five Smallest Neighbourhoods")
liv.plot(ax=ax, facecolor='black')
smallest_places.plot(ax=ax, facecolor='red')
ctx.add_basemap(ax, crs=liv.crs, source="http://mt0.google.com/vt/lyrs=m&hl=en&x={x}&y={y}&z={z}", attribution="Google Roadmap" )
ax.axis('off');
```

## Task III: The gender gap on the streets
`This one is a bit more advanced, so don’t despair if you can’t get it on your first try. It also relies on the streets dataset from the “Hands-on” section, so you will need to load it up on your own. Here’re the questions for you to answer:`

Which group accounts for longer total street length in Zaragoza: men or women? By how much?

```python
zaragoza = gpd.read_file(r"data/streets.geojson")
```

```python
# group by male and female then find total street length of each group.
zaragoza.columns
```

```python
zaragoza_gender_lengths = zaragoza.to_crs(epsg=25830).dissolve(by='gender').length.sort_values()
```

```python
f"The Male group accounts for longer total street length by {zaragoza_gender_lengths.iloc[-1]-zaragoza_gender_lengths[0]} metres"
```

### Second Method

```python
zaragoza_reprojected = zaragoza.to_crs(epsg=25830)
```

```python
male= zaragoza_reprojected[zaragoza_reprojected['gender']=='Male']
male.shape
```

```python
female = zaragoza_reprojected[zaragoza_reprojected['gender']=='Female']
female.shape
```

```python
male_length, female_length = sum(male.length), sum(female.length)
male_length, female_length
```

```python
f"The Male group accounts for longer total street length by {male_length - female_length} metres"
```

# Conclusion

```python
f"There is a difference of {abs((zaragoza_gender_lengths.iloc[-1]-zaragoza_gender_lengths[0]) - (male_length - female_length))} metres between the two results."
```
